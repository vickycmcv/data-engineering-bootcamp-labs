# Q. 11 — Practical: Simulate Batch and Streaming Events

Write one Python script that:

* Creates five order events.
* Prints all five together as a batch.
* Prints the same five one at a time with a short delay.
* Labels both sections clearly.

Expected Observation:

The batch section appears together. The streaming simulation appears gradually.

```python
import time
```

```python
order_events = [
    {"order_id":101,"amount":55.5},
    {"order_id":102,"amount":25},
    {"order_id":103,"amount":199.99},
    {"order_id":104,"amount":89.90},
    {"order_id":105,"amount":45.5}
]
```

```python
# Batch
for order in order_events:
    print(order)
```

```python
# Streaming
for order in order_events:
    time.sleep(1)
    print(order)
```

# Q. 12 — Practical: Ingest a Customer File into Raw Storage
BritBooks receives customers.csv.

### Input File:
<code>customer_id,name,email,city  
1, Amelia Clarke ,AMELIA@EXAMPLE.COM,Manchester  
2,Oliver Bennett,oliver@example.com,Bristol</code>  


### Task:
* Create data/source/customers.csv with the supplied data.
* Copy it unchanged to data/raw/customers.csv.
* Create missing folders.
* Print the number of copied data rows.

### Expected Output
<code>Raw ingestion complete: 2 rows</code>


```python
from pathlib import Path
import csv
```

```python
# Folders creation
source_path = Path("D:/Nextcloud/vcortes/Vicky/Cursos/Ironhack DE Bootcamp/Activities/Week3/Session1/data/source")
source_path.mkdir(exist_ok=True, parents=True)
raw_path = Path("D:/Nextcloud/vcortes/Vicky/Cursos/Ironhack DE Bootcamp/Activities/Week3/Session1/data/raw")
raw_path.mkdir(exist_ok=True, parents=True)
```

```python
# Create customer CSV in source
data= [
    ["customer_id", "name", "email", "city"],
    [1," Amelia Clarke ", "AMELIA@EXAMPLE.COM", "Manchester"],
    [2, "Oliver Bennett", "oliver@example.com", "Bristol"]
    ]
customer_source_path = source_path / "customers.csv"
with open(customer_source_path, "w" , encoding="utf-8") as file:
        writer = csv.writer(file, lineterminator='\n')
        writer.writerows(data)
```

```python
# Copy customer.csv from source to raw
customer_raw_path = raw_path / "customers.csv"
with open(customer_source_path, "r" ) as source_file:
        reader = csv.reader(source_file, delimiter=",")
        with open(customer_raw_path, "w" , newline='') as raw_file:
            writer = csv.writer(raw_file)
            for row in reader:
                writer.writerow(row)
```

```python
# Copy customer.csv from source to raw with shutil
# import shutil
# shutil.copyfile(customer_source_path, customer_raw_path)
```

```python
with open(customer_raw_path, "r" , encoding="utf-8") as raw_file:
        reader = csv.reader(raw_file)
        print(f"Raw ingestion complete: {sum(1 for row in reader)} rows")
```

Create a Python runner with these functions:

* <code>publish()</code> runs only when earlier functions succeed.

```python
def publish():
    print("Publishing was successful")
```

```python
steps = [ingest, clean, publish]
for step in steps:
    try:
        step()
    except Exception as error:
        print(f"Pipeline stopped at {step}: {error}")
        break
else:
    print("Pipeline finished successfully")    
```
